```python
import pandas as pd
deliveries=pd.read_csv('/content/drive/MyDrive/deliveries.csv')
deliveries
```

```python
total_score=deliveries.groupby(["match_id","inning"]).sum()["total_runs"].reset_index() #The .reset_index() function converts these indexes back into regular columns of the DataFrame.
```

```python
total_score=total_score[total_score["inning"]==1] # This condition will return True for rows where the inning is 1 and False otherwise.
```

```python
total_score
```

```python
 matches_df = matches.merge(total_score[["match_id",'total_runs']],left_on="id",right_on="match_id") #The merge is essentially matching each match (based on id) with the corresponding total runs for that match's first innings from total_score.
```

```python
matches_df.head()
```

```python
matches_df["team1"].unique()
```

```python
teams = [ 'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals', 'Chennai Super Kings',
        'Sunrisers Hyderabad',
        'Delhi Capitals', 'Punjab Kings',
       'Lucknow Super Giants', 'Gujarat Titans',
       'Royal Challengers Bengaluru']
#current playing teams
```

```python
matches_df["team1"]=matches_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals") #changing names
matches_df["team2"]=matches_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")
matches_df["team1"]=matches_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_df["team2"]=matches_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_df["team1"]=matches_df["team1"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")
matches_df["team2"]=matches_df["team2"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")
```

```python
matches_df=matches_df[matches_df["team1"].isin(teams)]
matches_df=matches_df[matches_df["team2"].isin(teams)]
```

```python
matches_df.shape
```

```python
matches_df.columns
```

```python
matches_df=matches_df[["match_id","city","winner","total_runs"]]
```

```python
matches_df.head()
```

```python
delivery_df = matches_df.merge(deliveries,on="match_id")
```

```python
delivery_df.head()
```

```python
delivery_df= delivery_df[delivery_df["inning"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df.head()
```

```python
delivery_df["total_runs_y"] = delivery_df["total_runs_y"].astype(int) #total runs already there so we are appending y so that both get differ
#here the runs changes into  int datatype ie "4" >> 4,"3">>3
```

```python
delivery_df.groupby(["match_id"])["total_runs_y"].cumsum() #The .cumsum() function calculates the cumulative sum of the total_runs_y values for each group (i.e., for each match_id).
#The cumulative sum means that, for each delivery in a match, it sums up all the runs scored from the start of the match up to that specific delivery.
```

```python
delivery_df["current_score"]=delivery_df.groupby(["match_id"])["total_runs_y"].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df["runs_left"]= delivery_df["total_runs_x"]-delivery_df["current_score"]
```

```python
delivery_df.head()
```

```python
delivery_df["balls_left"]=120-(delivery_df["over"]*6+delivery_df["ball"])
#EXAMPLE:If 3 overs are completed, and itâ€™s the 2nd ball of the 4th over, the total number of balls bowled is (3Ã—6)+2=18+2=20 balls.then balls left is 120-20 =100

```

```python
delivery_df.columns
```

```python
delivery_df["is_wicket"].dtype
```

```python
delivery_df["wickets_left"]= 10 - delivery_df.groupby(["match_id"])["is_wicket"].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df["crr"]= (delivery_df["current_score"]/(120-delivery_df["balls_left"]))*6 #curreentrunrate
#For 60 runs and 90 balls left:= 120 - 90 = 30 balls bowled
#crr=(60/30)*6=12 so crr is 12 runs per over
```

```python
delivery_df.head()
```

```python
delivery_df["rr"]= (delivery_df["runs_left"]/(delivery_df["balls_left"]))*6 #required runrate
#runsleft=20 and balls left=12 then  [rr=runsleft/ballsleft ](20/12)*6  so required runrarte is 10 per over
```

```python
def result (row):
    return 1 if row["batting_team"]==row["winner"] else 0
#The function returns 1 if the team in row["batting_team"] is the same as the team in row["winner"], indicating that the batting team won.
#Otherwise, it returns 0, indicating that the batting team did not win.
```

```python
delivery_df["result"]=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df= delivery_df[["batting_team","bowling_team","city","runs_left","balls_left","wickets_left","total_runs_x","crr","rr","result"]]
```

```python
final_df=final_df.sample(final_df.shape[0]) #randomly shuffling the rows of the final_df. shuffling before  spliting and testing helps us to overcome the order bais
```

```python
final_df.isnull()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True) #Dropping null values
```

```python
final_df=final_df[final_df["balls_left"]!=0] #Removing rows where balls left is equal to zero since that will lead to inf value error , as we were dividing using it while calculating crr and rrr .
```

```python
from sklearn.model_selection import train_test_split
X = final_df.iloc[:,:-1]#input feautures
y= final_df.iloc[:,-1]#taget variable
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=1)
#test_size=0.2: Specifies that 20% of the data should be used for testing, and the remaining 80% will be used for training.
#random_state=1: Sets the seed for the random number generator, ensuring that the split is reproducible.
#This means that each time you run the code with this random state, you will get the same split of data.
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
onehotencoder = ColumnTransformer([("onehotencoder",OneHotEncoder(sparse_output= False , drop="first"),["batting_team","bowling_team","city"])],remainder="passthrough")
```

```python
#MODEL SELECTION
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[
    ("step1",onehotencoder),
    ("step2",LogisticRegression(solver='liblinear'))])
pipe.fit(X_train  , y_train )
y_pred = pipe.predict(X_test)

#fit: This method is used to train the pipeline. It applies each step in the pipeline sequentially:
#The onehotencoder transforms the X_train data.
#The transformed data is then used by the LogisticRegression model to learn from y_train.

```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
%%writefile app.py


import pandas as pd
import streamlit as st
import pickle
import matplotlib.pyplot as plt

# Load the trained model (pipe.pkl)
pipe = pickle.load(open('pipe.pkl', 'rb'))

# Teams and Venues
teams = [
    'Mumbai Indians', 'Kolkata Knight Riders', 'Rajasthan Royals',
    'Chennai Super Kings', 'Sunrisers Hyderabad', 'Delhi Capitals',
    'Punjab Kings', 'Lucknow Super Giants', 'Gujarat Titans',
    'Royal Challengers Bengaluru'
]

venues = [
    'Bangalore', 'Delhi', 'Mumbai', 'Kolkata', 'Hyderabad', 'Chennai',
    'Jaipur', 'Cape Town', 'Port Elizabeth', 'Durban', 'Centurion',
    'East London', 'Johannesburg', 'Kimberley', 'Bloemfontein',
    'Ahmedabad', 'Cuttack', 'Nagpur', 'Visakhapatnam', 'Pune',
    'Raipur', 'Ranchi', 'Abu Dhabi', 'Bengaluru', 'Dubai',
    'Sharjah', 'Navi Mumbai', 'Chandigarh', 'Lucknow', 'Guwahati',
    'Dharamsala', 'Mohali'
]

# Dictionary for team colors
team_colors = {
    'Mumbai Indians': '#004BA0',  # Blue
    'Kolkata Knight Riders': '#3A225D',  # Purple
    'Rajasthan Royals': '#EA1A8E',  # Pink
    'Chennai Super Kings': '#FFFF00',  # Yellow
    'Sunrisers Hyderabad': '#FF822A',  # Orange
    'Delhi Capitals': '#17448A',  # Dark Blue
    'Punjab Kings': '#D71920',  # Red
    'Lucknow Super Giants': '#00A651',  # Green
    'Gujarat Titans': '#1C1C1C',  # Dark Grey
    'Royal Challengers Bengaluru': '#DA1D37',  # Red
}

# Dictionary for team abbreviations
team_abbr = {
    'Mumbai Indians': 'MI',
    'Kolkata Knight Riders': 'KKR',
    'Rajasthan Royals': 'RR',
    'Chennai Super Kings': 'CSK',
    'Sunrisers Hyderabad': 'SRH',
    'Delhi Capitals': 'DC',
    'Punjab Kings': 'PBKS',
    'Lucknow Super Giants': 'LSG',
    'Gujarat Titans': 'GT',
    'Royal Challengers Bengaluru': 'RCB'
}

# Custom CSS for styling
st.markdown(
    """
    <style>
    .main {
        background-color: #1E1E1E;
        border-radius: 15px;
        padding: 10px;
    }
    .prediction {
        font-size: 24px;
        font-weight: bold;
        color: white;
        text-align: center;
        margin-top: 20px;
    }
    .button {
        display: flex;
        justify-content: center;
        margin-top: 10px;
    }
    .stButton button {
        background-color: #007bff;
        color: white;
        border-radius: 8px;
        padding: 10px 30px;
        font-size: 20px;
    }
    </style>
    """,
    unsafe_allow_html=True,
)

# Main title
st.title("Predict - IPL Match Prediction Model")

# Layout for team and venue selection
col1, col2 = st.columns(2)
with col1:
    batting_team = st.selectbox("Select the Batting Team", sorted(teams))
with col2:
    bowling_team = st.selectbox("Select the Bowling Team", sorted(teams))

venue = st.selectbox("Select Venue", sorted(venues))

# Target and match details
target = st.number_input("Target")
col3, col4, col5 = st.columns(3)
with col3:
    score = st.number_input("Score")
with col4:
    overs = st.number_input("Overs completed")
with col5:
    wickets = st.number_input("Wickets out")

# Predict button
if st.button('Predict'):
    runs_left = target - score
    balls_left = 120 - (overs * 6)
    wickets_left = 10 - wickets
    crr = score / overs if overs > 0 else 0
    rr = (runs_left * 6) / balls_left if balls_left > 0 else 0

    input_df = pd.DataFrame({
        'batting_team': [batting_team], 'bowling_team': [bowling_team],
        'city': [venue], 'runs_left': [runs_left], 'balls_left': [balls_left],
        'wickets_left': [wickets_left], 'total_runs_x': [target],
        'crr': [crr], 'rr': [rr]
    })

    result = pipe.predict_proba(input_df)
    loss = result[0][0]
    win = result[0][1]

    # Get team colors and abbreviations
    batting_color = team_colors.get(batting_team, '#FFFFFF')  # Default to white if not found
    bowling_color = team_colors.get(bowling_team, '#FFFFFF')  # Default to white if not found
    batting_abbr = team_abbr.get(batting_team, 'Batting')
    bowling_abbr = team_abbr.get(bowling_team, 'Bowling')

    # Display the prediction result as text
    st.markdown("<div class='prediction'>{}</div>".format(
        f"{batting_team}: {round(win * 100)}% chance to win"
    ), unsafe_allow_html=True)
    st.markdown("<div class='prediction'>{}</div>".format(
        f"{bowling_team}: {round(loss * 100)}% chance to win"
    ), unsafe_allow_html=True)

    # Plotting stacked horizontal bar chart for win and loss probabilities
    fig, ax = plt.subplots(figsize=(10, 3))  # Adjusted figure size
    ax.barh([f'{batting_abbr} vs {bowling_abbr}'], [win * 100], color=batting_color)  # Color for batting team
    ax.barh([f'{batting_abbr} vs {bowling_abbr}'], [loss * 100], left=[win * 100], color=bowling_color)  # Color for bowling team

    # Remove the axes and spines
    ax.axis('off')

    # Set the background color to match the app's dark theme
    fig.patch.set_facecolor('#1E1E1E')  # Background color of the entire figure
    ax.set_facecolor('#1E1E1E')  # Background color of the plot area

    # Adding percentage text to the bar chart
    ax.text(win * 50, 0, f'{batting_abbr}: {int(win * 100)}%', va='center', ha='center', color='white', fontsize=12)
    ax.text(100 - (loss * 50), 0, f'{bowling_abbr}: {int(loss * 100)}%', va='center', ha='center', color='white', fontsize=12)

    st.pyplot(fig)

# Add some space at the bottom
st.markdown("<div class='button'></div>", unsafe_allow_html=True)


















```
